```python
# importing for tokenization
from tensorflow.keras.layers import TextVectorization
```

```python
# splitting comment text and rest of features into x and y respectively
x = df['comment_text']
y = df[df.columns[2:]].values
```

```python
# printing x
# optional
x
```

```python
# printing y
# optional
y
```

```python
# No of words
MAX_WORDS=200000
```

```python
# MAX_WORDS -> represents maximum words in the vocabulary
# output_sequence_length -> maximum length of a sentence. Will be more clear in cell no 56
vectorizer = TextVectorization(max_tokens=MAX_WORDS, output_sequence_length=1800, output_mode='int')
```

```python
# passing x(comment_text) passing to vectorizer for mapping words
vectorizer.adapt(x.values)
```

```python
# shows the result of mapping done in the previous line
# optional
vectorizer.get_vocabulary()
```

```python
# showing the values of random words
vectorizer('This will return values of each word')[:7]
```

```python
# now lets tokenize the our actual dataset using vectorizer
vectorized_text = vectorizer(x.values)
```

```python
# showing vectorized dataset
# optional
# note shape=(159571, 1800). That means 15971 sentences. Each sentence can have upto 1800 words. If sentence is not that long 0 is filled in its position
vectorized_text
```

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16)
dataset = dataset.prefetch(8)
```

```python
dataset.as_numpy_iterator().next()
```

```python
# taking 70% as training data
# taking next 20% as validation data
# taking 10% as testing data
# .skip makes sure we skip the data we already took
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*0.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*0.1))
```
